```python
import pandas as pd
```

```python
datasetpath='/content/drive/MyDrive/matches_1930_2022.csv'
```

```python
df=pd.read_csv(datasetpath)
```

```python
df
```

```python
def AnalyzeFrance():
  matchesPlayed = df['home_team'].value_counts()['France']
  print(f'France has played {matchesPlayed}')
```

```python
AnalyzeFrance()
```

```python
def AnalyzeTowns():
  TownData=df['home_team'].value_counts()
  for _ in TownData.index:
    print(f'{_} is {TownData[_]}')
```

```python
AnalyzeTowns()
```

```python
%matplotlib inline
```

```python
townName=[]
matchesPlayed=[]
TownData=df['home_team'].value_counts()
for _ in TownData.index:
  if TownData[_] > 30:
    townName.append(_)
    matchesPlayed.append(TownData[_])
```

```python
townName
```

```python
from matplotlib import pyplot as plt
```

```python
plt.bar(townName,matchesPlayed,)
plt.xlabel('Town Name')
plt.ylabel('Matches Played')
plt.title('Matches Played by Town')
plt.show()
```

```python
def matchesPlayedByTown():
  townName=[]
  matchesPlayed=[]
  TownData=df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30:
      townName.append(_)
      matchesPlayed.append(TownData[_])
  plt.bar(townName,matchesPlayed,)
  plt.xlabel('Town Name')
  plt.ylabel('Matches Played')
  plt.title('Matches Played by Town')
  plt.show()
```

```python
matchesPlayedByTown()
```

```python
def matchesPlayedByTown2():
  Above30=df['home_team'].value_counts()
  filteredData=Above30[Above30 > 30]
  filteredData.plot(kind='barh')
```

```python
matchesPlayedByTown2()
```
