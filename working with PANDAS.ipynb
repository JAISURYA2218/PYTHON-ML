```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')
```

```python
df['Discount'].fillna(0, inplace=True)
```

```python
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
filtered_df = df[df['TotalSale'] >= 20]
```

```python
filtered_df['PaymentType'] = filtered_df['PaymentType'].astype('category')
```

```python
category_sales = filtered_df.groupby('ProductCategory').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalSales=('TotalSale', 'sum')
).reset_index()
```

```python
category_sales_sorted = category_sales.sort_values(by='TotalSales', ascending=False)
```

```python
print("\nTop 3 Product Categories with Highest Total Sales:")
print(category_sales_sorted.head(3))
```

```python
store_product_avg_sales = filtered_df.groupby(['StoreID', 'ProductCategory'])['TotalSale'].mean().reset_index()
```

```python
top_category_per_store = store_product_avg_sales.loc[store_product_avg_sales.groupby('StoreID')['TotalSale'].idxmax()]
```

```python
top_category_per_store.rename(columns={'TotalSale': 'AvgTotalSale'}, inplace=True)
```

```python
print("\nStore-Level Insights:")
print(top_category_per_store[['StoreID', 'ProductCategory', 'AvgTotalSale']])
```
