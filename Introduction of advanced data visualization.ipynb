```python
f2=filteredData[filteredData>=40].index
```

```python
df2=df[df['home_team'].isin(f2)]
```

```python
df2['home_xg'].value_counts()
```

```python
%matplotlib inline
```

```python
df2.plot(x='home_team',y='home_xg',kind='bar')
```

```python
f3=df2['home_xg']>2.5
```

```python
df3=df2[f3]
```

```python
df3
```

```python
df3.plot(x='home_team',y='home_xg',kind='bar')
```

```python
from matplotlib import pyplot as plt
```

```python
df3.plot(x='home_team',y='home_xg',kind='bar')
plt.xlabel('Home Team')
plt.ylabel('Home XG')
plt.title('Home Team XG Stats')
```

```python
df3.plot(x='home_team',y=['home_score','home_xg'],kind='bar',color=['skyblue','pink'])
plt.xlabel('Home Team')
plt.ylabel('Home XG')
plt.title('Home Team estimated goals')
plt.show()
```
